**Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Data**

image.png

```python
X = np.array([30,37,25,43,50,29,46])
y = np.array([800,950,600,1050,1200,740,1100])
```

**Plot the data using matplotlib**

```python
plt.scatter(X,y)
```

**Implementing the Model using Gradient Descent**

image.png

```python
# Building the model
theta1 = 300
theta2 = 10
learning_rate = 0.0001

epochs = 10  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent
for i in range(epochs):
    y_pred = theta1 + theta2*X   # The current predicted value of Y
    D_theta2 = (2/n) * sum(X * (y_pred - y))  # Derivative wrt theta2
    D_theta1 = (2/n) * sum(y_pred - y)  # Derivative wrt theta1
    theta2 = theta2 - learning_rate * D_theta2  # Update theta2
    theta1 = theta1 - learning_rate * D_theta1  # Update theta1
    print (theta1, theta2)
```

**Compute the predicted values of given data using the learned model**
<br>Note: It is generally recommended to split the input data into **training and test sets**, allowing the model to be trained on the training set and evaluated on the test set. However, since the goal here is to focus on understanding how linear regression works and how gradient descent is used to learn the parameters, the data split has been omitted in this example.

```python
predictions = theta1 + theta2*X
predictions
```

**Plot the regression line over the input data.**

```python
plt.scatter(X, y)
plt.plot(X, predictions, color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear regression')
plt.show()
```
